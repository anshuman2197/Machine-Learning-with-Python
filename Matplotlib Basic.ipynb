# Histogram

```python
ml=np.random.randint(18,45,(100))
py=np.random.randint(15,440,(100))
print(ml,py)
```

```python
plt.hist(ml)
plt.title('ML Student')
plt.xlabel("Student age")
plt.ylabel("No of student")
plt.show()
```

```python
bins=[15,20,30,40,50]

plt.hist(ml,bins,rwidth=0.8)
plt.title('ML Student')
plt.xlabel("Student age")
plt.ylabel("No of student")
plt.show()
```

```python
bins=[15,20,30,40,50]

plt.hist(ml,bins,rwidth=0.8,histtype='step')
plt.title('ML Student')
plt.xlabel("Student age")
plt.ylabel("No of student")
plt.show()
```

```python
bins=[15,20,30,40,50]

plt.hist(ml,bins,rwidth=0.8,orientation='horizontal')
plt.title('ML Student')
plt.xlabel("Student age")
plt.ylabel("No of student")
plt.show()
```

```python
bins=[15,20,30,40,50]

plt.hist(ml,bins,rwidth=0.8,color='red')
plt.hist(py,bins,rwidth=0.8,color='green')

plt.title('ML Student')
plt.xlabel("Student age")
plt.ylabel("No of student")
plt.show()
```

```python
bins=[15,20,30,40,50]

plt.hist([ml,py],bins,rwidth=0.8,color=['red','green'],label=['ml','py'])
# plt.hist(py,bins,rwidth=0.8,color='green')

plt.title('Student')
plt.xlabel("Student age")
plt.ylabel("No of student")
plt.legend()
plt.show()
```

# Bar chart

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as style
```

```python
classes=["Python","R","AI","ML","DS"]
cl1=[30,10,20,25,10]
cl2=[40,5,20,20,10]
cl3=[35,5,30,15,15]
```

```python
plt.bar(classes,cl1)
```

```python
plt.barh(classes,cl2)
```

```python
plt.bar(classes,cl2,width=0.5, align='edge',color='y', edgecolor='m')
```

```python
plt.bar(classes,cl2,width=0.5, align='edge',color='y', edgecolor='m',linewidth=5)
```

```python
plt.bar(classes,cl2,width=0.5, align='edge',color='y', edgecolor='m',linewidth=5,alpha=0.9)
```

```python
plt.bar(classes,cl2,width=0.5, align='edge',color='y', edgecolor='m',linewidth=5,alpha=0.9, linestyle='--')
```

```python
plt.bar(classes,cl2,width=0.5, align='edge',color='y', edgecolor='m',linewidth=5,alpha=0.9, linestyle='--',label='class 3',visible=False)
```

```python
plt.style.use("ggplot")
plt.figure(figsize=(8,5))
plt.bar(classes,cl2,width=0.5, align='edge',color='y', edgecolor='m',linewidth=5,alpha=0.9, linestyle='--',label='class 3')
```

```python
plt.figure(figsize=(16,8))
cla=np.arange(len(classes))

width=0.2

plt.bar(cla,cl1,width,color="blue",label="class 1 student")

plt.bar(cla+width, cl2,width,color='green', label="class 2 student")
plt.bar(cla+width+width, cl3,width,color='orange', label="class 2 student")
plt.xticks(cla+width,classes,rotation=20)
plt.show()
```

# Pie Chart

```python
classes=['Python', 'R', 'AI', 'ML', 'DS']
cl1=[45,15,35,25,30]

explode=[0.03,0,0.1,0,0]
colors=["c","b","r","y","g"]
textprops={"fontsize":15}

plt.pie(cl1,labels=classes,explode=explode,colors=colors,autopct="%0.2f%%",shadow=True,radius=1.4,startangle=270,textprops=textprops)
plt.legend()
plt.show()
```
